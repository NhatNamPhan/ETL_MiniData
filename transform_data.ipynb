## Data cleaning and transform

```python
#Remove unnecessary columns and rename column
df = df.drop(['LongName','playerUrl','photoUrl','POT', 'Loan Date End'],axis=1)
df = df.rename(columns={'↓OVA':'OVA'})
```

```python
#Remove downstream characters 
df['Club'] = df['Club'].str.replace('\n\n\n\n','')
df['Club'].head()
```

```python
df['Height'] = df['Height'].str.strip("cm")
df['Height'].unique()
```

```python
#Convert ft to cm
def ft_to_cm(x):
    if "'" in x:
        ft = x.replace("'",".").replace('"','')
        cm = round(float(ft)*30.48,0)
        return int(cm)
    else:
        return int(x)
        
df['Height'] = df['Height'].apply(ft_to_cm)
df['Height'].unique()
```

```python
df['Weight'] = df['Weight'].str.strip("kg")
df['Weight'].unique()
```

```python
#Convert lbs to kg
def lbs_to_kg(x):
    if 'lbs' in x:
        lbs = x.replace('lbs','')
        kg = round(float(lbs)*0.4356)
        return int(kg)
    else:
        return int(x)
    
df['Weight'] = df['Weight'].apply(lbs_to_kg)
df['Weight'].unique()
```
